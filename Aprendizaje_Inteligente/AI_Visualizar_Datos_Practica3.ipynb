#### Practica No. 3 Visualizacion de Datos

#### Librerias


```python
import matplotlib.pyplot as plt
import numpy as np
```

##### Datos del eje x para los siguientes gr치ficos

```python
x = np.linspace(0, 10, 100)
print(x)
```

##### Ejemplos de gr치ficos de lineas (graficando funciones)

###### Nota: Lo siguiente se ejecuta todo junto

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

##### Paneles de Graficos

```python
plt.figure() # crear la figura
#Crear el primer plano
plt.subplot (2, 1, 1) # (filas, columnas, panel)
plt.plot(x, np.sin(x))
#Crear el segundo plano
plt.subplot (2, 1, 2)
plt.plot(x, np.cos(x))
```

##### Un estilo Orientado a Objetos para situaciones m치s complejas

```python
fig, ax = plt.subplots(2)
#Llama el metodo plot() method
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

##### Usando Colores (onda senoidal)

```python
plt.plot(x, np.sin(x - 0), color='blue') # Nombre del color
plt.plot(x, np.sin(x - 1), color='g') # codigo del color (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Escala de grises entre 0 y 1
#plt.plot(x, np.sin(x - 3), color='#FFDD44') # codigo hexadecimal (RRGGBB de 00 a FF)
#plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Tupla RGB de 0 a 1
#plt.plot(x, np.sin(x - 5), color='chartreuse') # soporta todos los nombres de HTML
```

##### Usando colores (lineas)

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

##### Usando Colores (lineas). Lo mismo pero con codigo

```python
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':') # dotted
```

##### Cambiamos los limites

```python
plt.plot(x,np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

##### Titulos

```python
plt.plot(x,np.sin(x))
plt.title("Funcion Seno")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

##### Leyendas

```python
plt.plot(x,np.sin(x), '-g', label='Seno(x)')
plt.plot(x,np.cos(x), ':b', label='Coseno(x)')
plt.legend(loc='upper right')
```

##### Ejemplos con los Datos de Iris

```python
import numpy as np
import pandas as pd

df = pd.read_csv("C:/Users/diego/Desktop/Universidad/6to-Semestre/Aprendizaje_Inteligente/iris.csv",delimiter=";",decimal=".")
print(df.shape)
print(df.head())
```

##### Elimina la variable categoria, deja las variables predictorias en X (Setosa y Versicolor)

```python
X = df.iloc[0:100, [0,2]].values
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o', label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='x', label='versicolor')

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')

plt.legend(loc='upper left')
plt.show()
```
